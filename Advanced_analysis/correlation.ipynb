```python
"""Relationships between Quantity and Value"""
import matplotlib.pyplot as plt
import pandas as pd

def correlation_analysis(df):
    # Calculate correlation
    correlation = df['QUANTITY'].corr(df['Total_Value'])
    print(f"Correlation between Quantity and Total Value: {correlation:.2f}")

    # Scatter plot
    plt.figure(figsize=(8, 6))
    plt.scatter(df['QUANTITY'], df['Total_Value'], alpha=0.6, color='teal')
    plt.title("Correlation between Quantity and Total Value")
    plt.xlabel("Quantity")
    plt.ylabel("Total Value")
    plt.tight_layout()
    plt.show()

# Perform correlation analysis
df = pd.read_csv('../cleaned_data.csv')
correlation_analysis(df)
```
